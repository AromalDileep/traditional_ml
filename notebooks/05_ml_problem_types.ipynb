# Supervised learning

### Supervised Learning — Brief Textbook Description

![Image](https://labelyourdata.com/cms/wp-content/uploads/2025/04/introduction-to-labeled-data-what-why-and-how_5.png)

![Image](https://media.geeksforgeeks.org/wp-content/uploads/20241022160725494723/supervised-machine-learning.webp)

![Image](https://builtin.com/sites/www.builtin.com/files/styles/ckeditor_optimize/public/inline-images/4_train-test-split.jpg)

**Supervised learning** is a machine learning approach in which a model is trained on a **labeled dataset**, meaning that each input is paired with a known output. The objective is to learn a function that maps input features to the correct target values, enabling the model to make accurate predictions on unseen data.

Formally, supervised learning seeks to approximate a function
$$
f(X) \rightarrow y
$$
where $X$ represents the input features and $y$ denotes the target variable.

Supervised learning problems are broadly classified into **regression**, where the target variable is continuous, and **classification**, where the target variable is categorical. During training, the model minimizes a **loss function** that quantifies the difference between predicted and actual outputs. Model performance is evaluated using appropriate **metrics** on a separate validation or test dataset to assess generalization.

Supervised learning is widely used in applications such as price prediction, medical diagnosis, image recognition, and spam detection due to its strong predictive performance when sufficient labeled data is available.


# Unsupervised Learning

### Unsupervised Learning — Brief Textbook Description

![Image](https://uk.mathworks.com/discovery/unsupervised-learning/_jcr_content/thumbnail.adapt.1200.medium.png/1730872445777.png)

![Image](https://storage.googleapis.com/static.prod.fiveable.me/search-images%2F%22Clustering_vs_dimensionality_reduction_in_unsupervised_learning_for_advanced_signal_processing_visualization_techniques%22-data_after_clustering.png)

![Image](https://media.geeksforgeeks.org/wp-content/uploads/20250526125548648108/What-is-Dimensionality-Reduction-.webp)

**Unsupervised learning** is a machine learning approach in which a model is trained on **unlabeled data**, meaning that no target variable or ground-truth output is provided. The objective is to **discover hidden patterns, structures, or relationships** inherent in the data without external supervision.

In unsupervised learning, the algorithm analyzes the input features (X) and organizes the data based on similarity, distribution, or statistical properties. Common tasks include **clustering**, where data points are grouped into similar subsets, **dimensionality reduction**, where the number of features is reduced while preserving important information, and **density estimation**, which models the underlying data distribution.

Unlike supervised learning, there is **no explicit loss function tied to known labels**. Model quality is often assessed using intrinsic measures such as cohesion, separation, variance preservation, or through downstream task performance. Unsupervised learning is widely used for exploratory data analysis, customer segmentation, anomaly detection, and feature learning, particularly when labeled data is scarce or expensive to obtain.

---

**One-line summary:**

> Unsupervised learning extracts structure and patterns from unlabeled data without predefined output labels.

Below is a concise, textbook-style list of **commonly used unsupervised learning algorithms**, grouped by task.

---

## 1. Clustering Algorithms

Used to group similar data points without labels.

* **K-Means Clustering**
  Partitions data into *K* clusters by minimizing within-cluster variance.

* **Hierarchical Clustering**
  Builds a tree (dendrogram) of clusters using agglomerative or divisive strategies.

* **DBSCAN (Density-Based Spatial Clustering)**
  Forms clusters based on density; identifies noise and outliers automatically.

* **Gaussian Mixture Models (GMM)**
  Assumes data is generated from a mixture of Gaussian distributions; probabilistic clustering.

---

## 2. Dimensionality Reduction Algorithms

Used to reduce feature space while preserving important information.

* **Principal Component Analysis (PCA)**
  Projects data onto orthogonal components with maximum variance.

* **t-SNE (t-Distributed Stochastic Neighbor Embedding)**
  Nonlinear technique mainly used for visualization in low dimensions.

* **UMAP (Uniform Manifold Approximation and Projection)**
  Preserves local and global structure; faster and more scalable than t-SNE.

---

## 3. Association Rule Learning

Used to discover relationships between variables.

* **Apriori Algorithm**
  Finds frequent itemsets and association rules in transactional data.

* **FP-Growth**
  Efficient alternative to Apriori using a tree-based structure.

---

## 4. Anomaly / Outlier Detection

Used to detect rare or abnormal data points.

* **Isolation Forest**
* **One-Class SVM**
* **Local Outlier Factor (LOF)**

---

## 5. Representation Learning

Learns compact feature representations.

* **Autoencoders**
  Neural networks that learn latent representations by reconstructing input data.

---

### One-line summary

> Unsupervised algorithms aim to discover structure, similarity, or patterns in unlabeled data through clustering, dimensionality reduction, association mining, or anomaly detection.


# Regression Vs Classification

## Regression vs Classification

![Image](https://dezyre.gumlet.io/images/blog/classification-vs-regression-in-machine-learning/image_46985406721641926364441.png?dpr=2.6\&w=376)

![Image](https://blogs.sas.com/content/graphicallyspeaking/files/2018/02/WFKReg1.png)

![Image](https://www.researchgate.net/publication/267974549/figure/fig1/AS%3A669535849943053%401536641194886/A-binary-classification-problem-the-decision-boundary-separating-hyperplane-x.png)

![Image](https://machinelearningmastery.com/wp-content/uploads/2020/01/Scatter-Plot-of-Multi-Class-Classification-Dataset.png)

### Core Difference

| Aspect            | Regression        | Classification           |
| ----------------- | ----------------- | ------------------------ |
| Target variable   | Continuous        | Categorical              |
| Output nature     | Numeric value     | Class label              |
| Objective         | Predict magnitude | Predict class membership |
| Decision boundary | Not required      | Required                 |

---

### Definition

* **Regression** predicts a **continuous-valued output** such as price, temperature, or time.
* **Classification** predicts a **discrete class label**, such as yes/no, spam/not spam, or digit classes.

---

### Examples

* Regression:

  * House price prediction
  * Stock price estimation
* Classification:

  * Email spam detection
  * Disease diagnosis

---

### Model Output

* Regression model outputs a **real number**.
* Classification model outputs:

  * A **class label**, or
  * **Class probabilities** (e.g., logistic regression)

---

### Loss Functions

| Task           | Common Loss                         |
| -------------- | ----------------------------------- |
| Regression     | MSE, MAE, RMSE                      |
| Classification | Cross-entropy, Log loss, Hinge loss |

---

### Evaluation Metrics

| Task           | Metrics                                  |
| -------------- | ---------------------------------------- |
| Regression     | MAE, RMSE, R²                            |
| Classification | Accuracy, Precision, Recall, F1, ROC-AUC |

---

### Algorithms Used

| Regression              | Classification           |
| ----------------------- | ------------------------ |
| Linear Regression       | Logistic Regression      |
| Ridge / Lasso           | Naive Bayes              |
| SVR                     | SVM                      |
| Decision Tree Regressor | Decision Tree Classifier |
| Random Forest Regressor | Random Forest Classifier |

---

### Key Conceptual Difference

* Regression focuses on **error magnitude**.
* Classification focuses on **decision boundaries** and **class separation**.

---

### Important Clarification

Even if classification labels are encoded as numbers (e.g., 0 and 1), the problem remains **categorical**, not numeric.

---

### One-line interview answer

> Regression predicts continuous values, while classification predicts discrete class labels.


# Clustering vs Classification

## Clustering vs Classification

![Image](https://cdn.educba.com/academy/wp-content/uploads/2023/10/Classification-vs-Clustering.jpg)

![Image](https://bookdown.org/tpinto_home/Unsupervised-learning/kmeans.png)

![Image](https://upload.wikimedia.org/wikipedia/commons/9/97/Linear_Decision_Boundary.png)

![Image](https://cdn.dida.do/blog/20230707_sp_supervised_vs_unsupervised_learning/illustration-supervised-vs-unsupervised.png)

### Core Difference

| Aspect           | Clustering         | Classification           |
| ---------------- | ------------------ | ------------------------ |
| Learning type    | Unsupervised       | Supervised               |
| Labels available | No                 | Yes                      |
| Objective        | Discover structure | Assign predefined labels |
| Output           | Groups (clusters)  | Class labels             |

---

### Definition

* **Clustering** groups data points based on **similarity** without knowing class labels.
* **Classification** assigns data points to **predefined classes** using labeled examples.

---

### Nature of Classes

* Clusters are **data-driven** and **emerge from the data**.
* Classes are **human-defined** and **fixed beforehand**.

---

### Example

* Clustering:
  Group customers based on purchasing behavior.
* Classification:
  Predict whether a customer will churn (Yes/No).

---

### Algorithms Used

| Clustering              | Classification      |
| ----------------------- | ------------------- |
| K-Means                 | Logistic Regression |
| Hierarchical Clustering | Decision Tree       |
| DBSCAN                  | SVM                 |
| Gaussian Mixture Model  | Random Forest       |

---

### Evaluation

* Clustering:
  Silhouette score, Davies–Bouldin index, visual inspection.
* Classification:
  Accuracy, Precision, Recall, F1-score.

---

### Decision Boundary

* Clustering does **not rely on decision boundaries**.
* Classification explicitly learns **decision boundaries**.

---

### Key Insight

* Clustering answers: *“What natural groups exist?”*
* Classification answers: *“Which known category does this belong to?”*

---

### One-line interview answer

> Clustering discovers groups in unlabeled data, whereas classification assigns labels based on learned decision boundaries from labeled data.

---


# Batch Learning vs Online Learning

## Batch Learning vs Online Learning

![Image](https://futureskillsacademy.com/wp-content/uploads/2025/01/Batch-Learning-vs-Online-Learning-1.png)

![Image](https://hazelcast.com/wp-content/uploads/2024/02/Hazelcast-AI-Landscape.png)

![Image](https://miro.medium.com/1%2A-xtyed8bESp8bykUemD7yw.png)

### Core Difference

| Aspect             | Batch Learning                 | Online Learning           |
| ------------------ | ------------------------------ | ------------------------- |
| Data availability  | Entire dataset at once         | Data arrives sequentially |
| Model updates      | Trained once (or periodically) | Updated continuously      |
| Adaptability       | Low                            | High                      |
| Computational cost | High at training time          | Spread over time          |

---

### Definition

* **Batch Learning** trains a model using the **entire dataset at once**. Once trained, the model remains fixed until it is retrained on a new batch of data.
* **Online Learning** updates the model **incrementally**, learning from one data point or a small mini-batch at a time as new data arrives.

---

### Data Flow

* Batch learning assumes **static data**.
* Online learning assumes **streaming or evolving data**.

---

### Examples

* Batch learning:

  * Training a house price model using historical data.
  * Training image classifiers offline.
* Online learning:

  * Recommendation systems adapting to user behavior.
  * Fraud detection systems learning from live transactions.

---

### Algorithms and Techniques

| Batch Learning                  | Online Learning                |
| ------------------------------- | ------------------------------ |
| Linear Regression (closed-form) | Stochastic Gradient Descent    |
| Random Forest                   | Online Gradient Descent        |
| XGBoost (offline)               | Perceptron                     |
| Deep Learning (epoch-based)     | Adaptive learning rate methods |

---

### Handling Concept Drift

* Batch learning requires **full retraining** when data distribution changes.
* Online learning can adapt naturally to **concept drift**.

---

### Memory and Resource Usage

* Batch learning requires storing the full dataset.
* Online learning processes data sequentially with limited memory.

---

### Key Trade-off

* Batch learning emphasizes **stability and accuracy**.
* Online learning emphasizes **adaptability and scalability**.

---

### One-line interview answer

> Batch learning trains models on the entire dataset at once, while online learning continuously updates models as new data arrives.

---

### Final takeaway

Both approaches solve the same predictive tasks; the choice depends on **data velocity, system constraints, and the need for real-time adaptation**.

## The Big Picture: How ML Problem Types Fit Together

### Step 1: Are labels available?

This is the **first and most important question**.

| Question                  | Answer | Learning Type         |
| ------------------------- | ------ | --------------------- |
| Do we have target labels? | Yes    | Supervised Learning   |
| Do we have target labels? | No     | Unsupervised Learning |

---

## Supervised Learning Branch

### Step 2: What type of output is required?

| Output Type      | Problem        |
| ---------------- | -------------- |
| Continuous value | Regression     |
| Discrete class   | Classification |

**Examples**

* Regression → house price, temperature
* Classification → spam detection, disease diagnosis

---

### Step 3: How is the model trained over time?

| Training Mode          | Approach        |
| ---------------------- | --------------- |
| All data at once       | Batch Learning  |
| Continuous data stream | Online Learning |

This applies to **both regression and classification**.

---

## Unsupervised Learning Branch

### Step 2: What is the goal?

| Goal                 | Problem                  |
| -------------------- | ------------------------ |
| Group similar points | Clustering               |
| Reduce features      | Dimensionality Reduction |
| Find anomalies       | Outlier Detection        |

---

### Key Contrast: Clustering vs Classification

* Clustering → discovers **unknown groups**
* Classification → assigns **known labels**

  Learns from labeled data to predict known outputs.

* **Unsupervised Learning**
  Discovers structure in unlabeled data.

* **Regression vs Classification**
  Continuous values vs discrete labels.

* **Clustering vs Classification**
