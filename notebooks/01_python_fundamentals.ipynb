## Core Data Types & Containers

These are used constantly when handling data before it reaches NumPy/Pandas.

We focus on:
- `list`
- `tuple`
- `set`
- `dict`


```python
# List
features = ["age", "salary", "experience"]
features.append("education")

# Tuple (immutable)
shape = (100, 3)

# Set (unique values)
unique_labels = {0, 1, 1, 0}

# Dictionary
sample = {
    "age": 30,
    "salary": 50000,
    "target": 1
}

features, shape, unique_labels, sample
```

## Control Flow

Used for:
- Data inspection
- Conditional preprocessing
- Simple rule-based logic


```python
# Conditional
age = 25
if age >= 18:
    status = "adult"
else:
    status = "minor"

# Loop
values = [1, 2, 3, 4, 5]
squared = []

for v in values:
    squared.append(v ** 2)

status, squared
```

## Comprehensions

Comprehensions are preferred in ML code because they are:
- Concise
- Readable
- Efficient

They appear frequently in preprocessing logic.


```python
# List comprehension
squared_comp = [v ** 2 for v in values]

# Conditional comprehension
even_numbers = [v for v in values if v % 2 == 0]

# Dict comprehension
feature_lengths = {f: len(f) for f in features}

squared_comp, even_numbers, feature_lengths
```

## Functions

Functions are used to:
- Encapsulate logic
- Improve readability
- Avoid repetition

ML code relies heavily on small, well-defined functions.


```python
def normalize_min_max(values):
    min_val = min(values)
    max_val = max(values)
    return [(v - min_val) / (max_val - min_val) for v in values]

normalize_min_max([10, 20, 30, 40])
```

## Lambda Functions

Used for:
- Short, one-line operations
- Sorting
- Mapping simple transformations

They are common but should not replace readable functions.


```python
# Lambda for sorting
records = [
    {"name": "A", "score": 80},
    {"name": "B", "score": 95},
    {"name": "C", "score": 70}
]

sorted_records = sorted(records, key=lambda x: x["score"], reverse=True)
sorted_records
```

## zip & enumerate

These are frequently used for:
- Iterating over multiple sequences
- Index-aware loops


```python
# zip
names = ["A", "B", "C"]
scores = [80, 90, 85]

paired = list(zip(names, scores))

# enumerate
indexed = [(i, name) for i, name in enumerate(names)]

paired, indexed
```

## *args and **kwargs

Used in:
- Utility functions
- Flexible APIs
- Wrapper functions


```python
def mean_values(*args):
    print(type(args))
    return sum(args) / len(args)

def print_config(**kwargs):
    print(type(kwargs))
    return kwargs

mean_values(1, 2, 3, 4), print_config(model="logistic", lr=0.01)
```

## collections.Counter

Very useful for:
- Label distribution
- Frequency analysis
- Class imbalance inspection


```python
from collections import Counter

labels = [0, 1, 1, 0, 1, 1, 1]
label_counts = Counter(labels)

label_counts
```

## Key Takeaways

- Python is used for **glue logic** around ML, not heavy computation
- Clean data structures improve downstream NumPy/Pandas work
- Functions and comprehensions improve readability
- `Counter`, `zip`, and `enumerate` appear frequently in ML pipelines
