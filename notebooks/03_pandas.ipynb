# Handling missing values

`````python
df = sns.load_dataset('titanic')

df.isna().sum()
`````

`````python
df.dropna()
`````

`````python
# Drop where specific column is missing
df.dropna(subset=['age'])
`````

`````python
# drop column with missing values 
df.dropna(axis=1)
`````

`````python
# fill with a constant 
# df.fillna(0)
num_cols = df.select_dtypes(include="number")
df[num_cols.columns] = num_cols.fillna(0)
`````

`````python
# Fill with mean / median / mode
df["age"].fillna(df["age"].mean(), inplace=True)
`````

`````python
df["age"].fillna(df["age"].median(), inplace=True)
`````

`````python
df["age"].fillna(df["age"].mode(), inplace=True)
`````

`````python
# Forward fill and Backward fill
df.fillna(method="ffill")
`````

`````python
df.fillna(method="bfill")
`````

# grouby and agg

`````python
df = sns.load_dataset("titanic")
df.head()
`````

`````python
df.groupby("sex")["survived"].mean()
`````

1. df.groupby("sex")

- Returns a DataFrameGroupBy object

- No aggregation is computed yet

2. ["survived"]

- Selects a column

- Returns a SeriesGroupBy object

3. .mean()

- Triggers computation

- Produces a concrete Series

`````python
df.groupby("sex")
`````

- grouby() creates a GroupBy object, which is a lazy, iterable view over grouped data.

- #### What is a GroupBy object?

A GroupBy object is:

    Lazy (no computation until aggregation)

    Iterable (you can loop over it)

    Not a generator

`````python
# Groupby multiple columns 
df.groupby(["sex", "class"])["survived"].mean()
`````

`````python
# Using agg()

df.groupby('class').agg(
    avg_age = ("age", "mean"),
    max_age=("age", "max"),
    avg_fare=("fare", "mean"),
    count=("fare", "count")
)
`````

- Syntax 
```
df.groupby(group_cols).agg(
    output_col_name = (input_col_name, aggregation_function),
    ...
)
```


# Pandas GroupBy Aggregation – Cheat Sheet

## Basic Pattern
```python
df.groupby("group_col").agg(
    out_col=("source_col", "agg_func")
)
````

---

## Core Aggregations (Most Used)

* `mean`
* `sum`
* `min`
* `max`
* `count`   # excludes NaN
* `size`    # includes NaN
* `median`
* `std`
* `var`
* `nunique`

---

## Descriptive / Distribution

* `quantile`
* `idxmin`
* `idxmax`
* `sem`
* `prod`

--- 


* **`quantile(q)`** — returns the value below which *q × 100%* of the data falls
* **`idxmin`** — returns the index label of the minimum value in the group
* **`idxmax`** — returns the index label of the maximum value in the group
* **`sem`** — returns the standard error of the mean *(std / √n)*
* **`prod`** — returns the product of all values in the group


---

## Boolean

* `any`
* `all`

---

## Categorical / String

* `first`
* `last`
* `unique`
* `value_counts`

---

## Valid Function Types

```python
("age", "mean")        # pandas string
("age", np.mean)      # NumPy
("age", len)          # Python builtin
("age", lambda x: x.max() - x.min())  # custom
```

---

## Multiple Aggregations

```python
df.groupby("class").agg(
    avg_age=("age", "mean"),
    max_age=("age", "max"),
    cnt=("fare", "count")
)
```

---

## count vs size

| Function | Counts NaN |
| -------- | ---------- |
| count    | ❌ No       |
| size     | ✅ Yes      |

---

## Rules

* Each aggregation returns **one scalar**
* Output column names are **mandatory**
* Same column can be aggregated multiple times
* Computation is **lazy** until `.agg()` is called

---

## Mental Model

`groupby().agg()` = **SQL GROUP BY + SELECT**

```sql
SELECT class, AVG(age), MAX(age)
FROM df
GROUP BY class;
```

```


`````python
# Multiple aggregation on same column
df.groupby("sex")["age"].agg(["mean", "median", "min", "max"])
`````

`````python
# Rest Index (Flat DataFrame)
df.groupby("sex")["survived"].mean().reset_index()
`````

- reset_index() converts the group-by index (sex) back into a regular column, turning the result from an indexed Series into a standard DataFrame.

Mental model
| Step        | Meaning                     |
| ----------- | --------------------------- |

# Merging and Joining

`````python
customers = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"]
})

orders = pd.DataFrame({
    "customer_id": [1, 1, 2],
    "order_amount": [250, 450, 300]
})
`````

`````python
customers
`````

`````python
orders
`````

`````python
# # Merging 
pd.merge(customers, orders, on="customer_id")
# Only matching customer_ids
# # Drops customers with no orders
`````

`````python
# left join 
pd.merge(customers, orders, on="customer_id", how="left")
`````

`````python
# Right join
pd.merge(customers, orders, on="customer_id", how="right")
`````

`````python
#outer join 
pd.merge(customers, orders, on="customer_id", how="outer")
# Keeps all rows from both tables
`````

`````python
# Merge with different column names 
pd.merge(
    customers,
    orders,
    left_on="customer_id",
    right_on="customer_id"
)
# (Useful when column names differ.)
`````

# Join using index join()


`````python
customers.set_index("customer_id").join(
    orders.set_index("customer_id"),
    how="left"
)
# join() works on index
# merge() works on columns
`````

`````python
# Many to one merge(Very common) 

pd.merge(orders, customers, on="customer_id")

# Many orders -> one customer 
# No duplication issues if keys are correct
`````

`````python
# Check merge correctness 
pd.merge(customers, orders, on="customer_id", validate="one_to_many")
# Helps catch bug early 
`````

- merge() joins DataFrames using columns, while join() joins DataFrames using the index (often making the join key the index)
