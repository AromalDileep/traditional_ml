# linspace

```python
# linspace and Normalization
x = np.array([10, 20, 50, 80])
x_min, x_max = x.min(), x.max()

x_norm = (x - x_min) / (x_max - x_min)
grid = np.linspace(0, 1, len(x))

print("Normalized data:", x_norm)
print("Reference grid:", grid)

# Interpretation
# x_norm tells you where each value lies in the [0, 1] interval
# linspace defines the ideal evenly spaced reference positions

# This is exactly how:
# feature scaling
# histogram binning
# quantization
```

```python
# linspace and Interpolation
# Interpolation estimates values between known data points.
# linspace provides the query locations where interpolation happens.

# Known samples
x = np.array([0, 1, 2, 3])
y = np.array([0, 1, 4, 9])  # y = x^2

# Create dense query points
x_new = np.linspace(x.min(), x.max(), 20)

# Linear interpolation
y_new = np.interp(x_new, x, y)
```

## flatten
- flatten() converts a multi-dimensional array into a one-dimensional array by copying all elements into a new contiguous array.

```python

xs = np.array([[1, 2, 3],
              [4, 5, 6]])

ys = xs.flatten()
ys
```

# Ravel
- ravel() converts an array into a one-dimensional array by flattening it when possible without copying data.

```python
xs = np.array([[1, 2, 3],
              [4, 5, 6]])

ys = xs.ravel()

"""
flatten() → always makes a copy
ravel() → returns a view if possible, otherwise a copy
"""
ys
```

## Boolean Masking

Boolean masking allows selecting rows or values
based on conditions.

This is heavily used during data cleaning and preprocessing.


```python
# Condition on the first feature
mask = X[:, 0] > 0

mask[:10]
# mask is a Boolean array
# Same length as number of rows in X
```

```python
# Select rows where first feature > 0
X_positive = X[mask]

X_positive.shape
```

```python
# Multiple conditions
mask_multi = (X[:, 0] > 0) & (X[:, 1] < 0)

X_filtered = X[mask_multi]

X_filtered.shape
# It selects all rows form X where the first feature is positive and the
# second feature is negative
```

## Broadcasting

Broadcasting allows NumPy to perform operations on arrays
with different shapes by automatically expanding dimensions
when it is mathematically valid.

This is heavily used in:
- Feature scaling
- Normalization
- Centering data


```python
# Mean and standard deviation per feature (column-wise)
mean = X.mean(axis=0)
std = X.std(axis=0)

mean, std
```

```python
# Standardization using broadcasting
X_standardized = (X - mean) / std

X_standardized[:3]
# It standardizes each feature (column) of X so that 
# every feature has mean 0 and standard deviation 1
"""
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)
"""
```

```python
# After standardization:
X_standardized.mean(axis=0), X_standardized.std(axis=0)
```

## Aggregations and `axis`

`axis` tells NumPy **which direction to collapse**.

Rule of thumb:
- `axis=0` → column-wise (down the rows)
- `axis=1` → row-wise (across columns)


```python
# Column-wise mean (per feature)
X.mean(axis=0)
```

```python
# Row-wise mean (per sample)
X.mean(axis=1)[:5]
```

- Aggregation means reducing multiple values into a single summary value.

```python
# aggregation over axis
x = np.array([
    [1,2,3],
    [4,5,6]])

np.sum(x, axis=0)
```

```python
np.sum(x, axis=1)
```

- Aggregation reduces dimensionality

```python
x.shape               # (2, 3)
np.sum(x, axis=0).shape  # (3,)
np.sum(x, axis=1).shape  # (2,)
np.sum(x).shape          # ()
```

-  keepdims=True (critical for broadcasting)

```python
np.sum(x, axis=1, keepdims=True)
# Keeps dimensions for later broadcasting
```

```python
#  Multiple axes aggregation
x = np.random.rand(2, 3, 4)

np.mean(x, axis=(1, 2))
x
```

```python
np.mean(x)        # NaNs propagate
np.nanmean(x)     # Ignores NaNs
```

```python
x = np.array([True, False, True])

np.any(x)   # True 
np.all(x)   # False

```

# Dot product
- Give weighted sum of features

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

np.dot(a, b)
# or
a @ b
print(a@b)
```

- The dot product of two vectors is the sum of pairwise component multiplications, equal geometrically to the product of their magnitudes and the cosine of the angle between them.

-  np.dot() — legacy, overloaded behavior


```python
np.dot([1,2,3],[4,5,6])
```

- np.matmul() — clean matrix multiplication
- @ operator — recommended syntax

- @ = np.matmul()
- np.dot() != np.matmul() -> only the same in some cases

```python
# When all three give the same result
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

a @ b
np.matmul(a, b)
np.dot(a, b)
```

```python
# When np.dot() behaves differently 
# Higher dimensional arrays 
A = np.random.rand(10, 2, 3)
B = np.random.rand(10, 3, 4)
A @ B
```

```python
np.dot(A,B)
```

- The @ operator is exactly equivalent to np.matmul(), while np.dot() only matches them for simple 1D or 2D cases and behaves differently for higher-dimensional arrays.

- Use @ → always
- Use matmul() → if you need function form
- Use dot() → only when reading old code


# simple normalization using numpy


```python
import numpy as np 

x = np.array([10,20,30,40])

x_min = x.min()
x_max = x.max()

x_norm = (x-x_min) / (x_max-x_min)
print(x_norm)
```

```python
# normalizing a 2d array

x= np.array([[1,10],
             [2,20],
             [3,30]])

x_min = x.min(axis=0)
x_max = x.max(axis=0)

x_norm = (x-x_min) / (x_max - x_min)
print(x_norm)
```

```python
# image style normalization 
image = np.array([0,128,255], dtype=np.float32)
image_norm = image / 255.0

print(image_norm)
```
