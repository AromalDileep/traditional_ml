### 1. Definition

**K-Means Clustering** is an **unsupervised machine learning algorithm** used to partition a dataset into **K distinct, non-overlapping clusters**.
Each cluster is represented by its **centroid**, and data points are assigned to the cluster whose centroid is closest to them.

The objective of K-Means is to **minimize intra-cluster variance**, also called **Within-Cluster Sum of Squares (WCSS)**.

---

### 2. Objective Function

K-Means minimizes the following cost function:

$$
J = \sum_{i=1}^{K} \sum_{x \in C_i} | x - \mu_i |^2
$$

Where:

* $K$ = number of clusters
* $C_i$ = set of points in cluster (i)
* $\mu_i$ = centroid (mean) of cluster (i)
* $||x - \mu_i||^2$ = squared Euclidean distance

---

### 3. Assumptions of K-Means

* Number of clusters $K$ is **predefined**
* Clusters are **spherical**
* All features are **numerical**
* Distance is usually **Euclidean**
* Clusters have **similar variance**

---

## 4. Step-by-Step Working of K-Means

### **Step 1: Choose the Number of Clusters (K)**

Decide the value of $K$ based on domain knowledge or methods like:

* Elbow Method
* Silhouette Score

---

### **Step 2: Initialize Centroids**

Randomly select **K data points** from the dataset as initial centroids.

> Poor initialization can lead to suboptimal clustering.

---

### **Step 3: Assign Data Points to Nearest Centroid**

For each data point:

1. Compute distance to each centroid
2. Assign the point to the cluster with **minimum distance**

Mathematically:

$$
Cluster(x) = \arg\min_j | x - \mu_j |
$$

---

### **Step 4: Update Centroids**

Recalculate each centroid as the **mean of all data points** assigned to that cluster:

$$
\mu_j = \frac{1}{|C_j|} \sum_{x \in C_j} x
$$

---

### **Step 5: Repeat Assignment and Update**

Repeat **Step 3** and **Step 4** until:

* Centroids no longer change, or
* Cluster assignments remain constant, or
* Maximum number of iterations is reached

---

### **Step 6: Convergence**

The algorithm converges when the objective function reaches a **local minimum**.

---

## 5. Termination Criteria

K-Means stops when **any one** of the following occurs:

* No change in centroids
* No change in cluster assignments
* Maximum iterations reached

---

## 6. Output of K-Means

* Final cluster centroids
* Cluster label for each data point
* Minimized WCSS value

---

## 7. Time and Space Complexity

* **Time Complexity:**
  $$
  O(n \cdot k \cdot d \cdot i)
  $$
  Where:

* $n$ = number of data points

* $k$ = number of clusters

* $d$ = number of features

* $i$ = number of iterations

* **Space Complexity:**
  $$
  O(n + k)
  $$

---

## 8. Advantages of K-Means

* Simple and easy to implement
* Computationally efficient
* Scales well to large datasets

---

## 9. Limitations of K-Means

* Must specify (K) beforehand
* Sensitive to initial centroids
* Sensitive to outliers
* Works poorly with non-spherical clusters
* Converges to **local minimum**, not global minimum

---

* K-Means is an unsupervised clustering algorithm
* It partitions data into K clusters by minimizing intra-cluster variance
* The algorithm iteratively assigns points and updates centroids
* Convergence occurs when centroids stabilize

### Step 1: Initialize Centroids

* Select **K random data points** as the **initial centroids**.
* These are temporary starting points.

---

### Step 2: Assign Data Points to Nearest Centroid

* For **each data point**, compute its distance to **each centroid**.
* Assign the data point to the cluster with the **minimum distance**.

> Distance is computed **from data points to centroids**, not between data points.

---

### Step 3: Update Centroids Using Mean

* For each cluster, compute the **mean of all assigned data points**.
* This **mean position becomes the new centroid**.

$$
\mu = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

---

### Step 4: Repeat

* Reassign data points using updated centroids.
* Recompute centroids using the new assignments.
* Continue until centroids **stop moving**.

---

## ❌ What Is NOT Done (Common Confusions)

### ❌ “Find nearest data point to the mean and make that the centroid”

* **Incorrect for K-Means**
* The centroid **does not have to be an actual data point**
* It can lie **anywhere in feature space**

> This idea belongs to **K-Medoids**, not K-Means.

---

### ❌ “Calculate distance from centroid to nearest data point first”

* Distances are always calculated **from each data point to all centroids**
* Assignment is **point → centroid**, never **centroid → point**

---

## Corrected Version of Your Explanation (Clean & Short)

> First, K centroids are initialized by randomly selecting data points.
> Then, each data point is assigned to the nearest centroid based on distance.
> After assignment, new centroids are computed as the mean of all points in each cluster.
> This process repeats until the centroids no longer change.

---

## One-Line Rule to Remember (Very Important)

**K-Means centroid = mean of points (not an actual data point).**
**Distance is computed from data points to centroids.**

## Elbow Method 

The graph is called Elbow curve
- X axis -> no. of clusters
- y axis -> WCSS(Within Clusters Sum of Squared distance) / Inertia

### WCSS

WCSS = We calculate the distance of data points in each cluster to its centroid and take the sum of the squares of those distances.

Final WCSS is obtained by summing the WCSS values of all clusters.


### Elbow point 
WCSS decreases as the number of clusters increases. When all data points belong to a single cluster, WCSS is high. As clusters increase, data points are closer to their centroids, reducing WCSS. When the number of clusters equals the number of data points, each point becomes its own cluster and WCSS becomes zero.

`The elbow point is the value of K where the reduction in WCSS is no longer drastic.`


# K-means (code) 

## Iris dataset

The value of K in K-Means is usually determined using the Elbow Method. Other approaches include the Silhouette Score and domain knowledge. GridSearchCV is generally not used since K-Means is an unsupervised algorithm.


```python
import seaborn as sns 

df = sns.load_dataset("iris")
df.head()
```

```python
df.info()
```

```python
X = df.drop(columns=["species"])
X.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

pipeline = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("kmeans",KMeans(random_state=42))
])
```

In K-Means, the number of clusters (K) is the primary hyperparameter that controls the clustering structure. Other parameters such as initialization method and n_init mainly improve stability and convergence rather than changing the objective. In practice, these are fixed to sensible defaults while K is selected using methods like the elbow or silhouette analysis.


### Elbow method

```python
import matplotlib.pyplot as plt 

wcss = []

for k in range(1,11):
    pipeline.set_params(kmeans__n_clusters=k)
    pipeline.fit(X)
    wcss.append(pipeline.named_steps["kmeans"].inertia_)

plt.plot(range(1,11),wcss, marker="o")
plt.xticks(range(1,11)) # force ticks 1,2,3
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method (Iris Dataset)")
plt.show()
```

### Silhouette Method

### Silhouette Plot

* **X-axis → Number of clusters (K)**
* **Y-axis → Silhouette Score**

The silhouette score measures **how well data points are clustered** by comparing **intra-cluster cohesion** and **inter-cluster separation**.

---

### Silhouette Score (Definition)

For each data point $i$:

1. **$a(i)$** → Average distance from point $i$ to all other points in the **same cluster**
   (measures how well the point fits within its own cluster)

2. **$b(i)$** → Minimum average distance from point $i$ to points in the **nearest neighboring cluster**
   (measures separation from other clusters)

The silhouette value for point $i$ is:

$$
s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}
$$

---

### Silhouette Score Range

$$
-1 \leq s(i) \leq 1
$$

* **Close to +1** → well-clustered
* **≈ 0** → overlapping clusters
* **Negative** → likely misclassified

---

### Overall Silhouette Score

The **final silhouette score** is the **average of silhouette values for all data points**.

$$
\text{Silhouette Score} = \frac{1}{n} \sum_{i=1}^{n} s(i)
$$

---

### Choosing K using Silhouette Method

* Compute silhouette score for different values of (K)
* Choose the value of **K with the highest silhouette score**

---

### Key Interpretation Rules

* **Higher silhouette score ⇒ better clustering**
* Penalizes:

  * Loose clusters
  * Overlapping clusters
* Works best when clusters are **compact and well separated**

---

### Comparison with Elbow Method

| Elbow Method         | Silhouette Method          |
| -------------------- | -------------------------- |
| Uses WCSS            | Uses cohesion & separation |
| Heuristic and visual | Quantitative metric        |
| Subjective           | Objective                  |
| Always decreases     | Has a clear maximum        |

---

### Important Insight (Very Exam-Relevant)

> Elbow method suggests candidate values for K, while silhouette score quantitatively validates the choice.

---

### One-Line Summary

Silhouette score evaluates clustering quality by comparing how close a data point is to its own cluster versus the nearest neighboring cluster, and the optimal K is chosen where the silhouette score is maximum.


```python
from sklearn.metrics import silhouette_score

sil_scores = []

for k in range(2, 11):
    pipeline.set_params(kmeans__n_clusters=k)
    pipeline.fit(X)
    labels = pipeline.named_steps["kmeans"].labels_
    sil_scores.append(silhouette_score(
        pipeline.named_steps["scaler"].transform(X),
        labels
    ))

plt.plot(range(2, 11), sil_scores, marker="o")
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs K")
plt.show()
```

### Improve

```python
pipeline.set_params(kmeans__n_clusters=3, kmeans__n_init=20)
pipeline.fit(X)

df["cluster"] = pipeline.named_steps["kmeans"].labels_

print(pd.crosstab(df["species"], df["cluster"]))
```

kmeans__n_init controls how many times K-Means is run with different random initial centroids.

• K-Means can converge to different solutions depending on initial centroid positions.
• n_init runs the algorithm multiple times with different initializations.
• The solution with the lowest WCSS (inertia) is selected as the final model.

Default value:
• In modern scikit-learn versions (≥ 1.4), the default is n_init = "auto"
• "auto" typically uses 10 initializations for random init
• In older versions, the default was n_init = 10

Why n_init = 20 is used:
• Increases robustness against poor initialization
• Reduces chance of converging to a bad local minimum
• Produces more stable and reliable clustering results

Summary:
• Higher n_init → better stability (but more computation)
• n_init does NOT change K or distance metric
• n_init affects only centroid initialization quality


```python
print(pd.crosstab(df["species"], df["cluster"]))
```

**Insight (concise and project-ready):**

* K-Means successfully **isolates the well-separated Setosa class**, which indicates the algorithm is capturing the strongest geometric structure in the data.
* **Versicolor and Virginica overlap in feature space**, so K-Means cannot separate them cleanly; their mixing is expected and not a failure.
* Increasing `n_init` improved **clustering stability** and reduced WCSS, even though label alignment did not improve uniformly.
* This result highlights a key principle: **K-Means optimizes distance (compactness), not class purity**.
* Therefore, clustering quality should be judged using **internal metrics (WCSS, silhouette)** and domain understanding, not direct label matching.

This is a **correct and defensible outcome** for the Iris dataset in an unsupervised setting.

# Silhouette (validation)

The silhouette coefficient or silhouette score is a measure of how similar a data point is within-cluster (cohesion) compared to other clusters (separation)

```python
from sklearn.metrics import silhouette_score

sil_scores = []

for k in range(2,11):
    pipeline.set_params(kmeans__n_clusters=k)
    pipeline.fit(df)

    labels = pipeline.named_steps["kmeans"].labels_
    X_scaled = pipeline.named_steps["scaler"].transform(df)
    sil_scores.append(silhouette_score(X_scaled, labels))

plt.plot(range(2,11), sil_scores, marker="o")
plt.xticks(range(2,11))
plt.xlabel("Number of clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs K (Wine Dataset)")
plt.show()
```

### Clear Distinction: `labels_` vs Silhouette Value $ s(i) $

#### 1. `labels_` (Cluster Assignment)
- `labels_` is a vector of **cluster IDs**
- It tells **which cluster each data point belongs to**
- Discrete and integer-valued

Example:
$$
\text{labels\_}[i] = k
$$

Meaning:
- Data point $ i $ is assigned to cluster $ k $

---

#### 2. Silhouette Value $ s(i) $ (Cluster Quality Measure)
- $ s(i) $ measures **how well a data point fits within its assigned cluster**
- Continuous value in the range $[-1, 1]$

Formula:
$$
s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}
$$

Where:
- $ a(i) $ = average distance of point $ i $ to **other points in the same cluster**
- $ b(i) $ = average distance of point $ i $ to **points in the nearest other cluster**

---

#### 3. How They Work Together
- `labels_` is used to identify cluster membership
- Cluster membership determines $ a(i) $ and $ b(i) $
- $ a(i) $ and $ b(i) $ are then used to compute $ s(i) $

---

#### Summary
- **`labels_` → cluster identity**
- **$ s(i) $ → quality of that assignment**
- `labels_` is an **input** to the silhouette formula
- $ s(i) $is the **output**
